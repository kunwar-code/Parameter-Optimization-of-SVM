```python
import pandas as pd
import numpy as np
import random

df= pd.read_csv('adult.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df=df.dropna(0,'any')
```

```python
df.isnull().sum()
```

```python
samples=[]
for i in range(10):
    samples.append(df.sample(n=3000))
    
```

```python
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn import preprocessing
```

```python
def fun(s_df):
    encoder = preprocessing.LabelEncoder()
    x=s_df.iloc[:,:14]
    x=pd.get_dummies(x)
    y=s_df.income
    encoder.fit(y)
    y = encoder.transform(y)
    x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.7, random_state = 0)
    kernellist=['rbf','poly','sigmoid','linear']
    bestAccuracy=0
    bestKernel=""
    bestC=0
    bestGamma=0
    
    for i in range(10):
        
        k=random.choice(kernellist)
        c=random.random()
        g=random.random()
        
        model=svm.SVC(kernel=k,C=c,gamma=g,decision_function_shape='ovo').fit(x_train,y_train)
        pred=model.predict(x_test)
        accuracy=model.score(x_test,y_test)
        
        if(accuracy>bestAccuracy):
            bestKernel=k
            bestC=c
            bestGamma=g
            bestAccuracy=accuracy
    
    
    return bestAccuracy,bestKernel,bestC,bestGamma

    
```

```python
result=[]
for s in samples:
    a,k,c,g=fun(s)
    result=[{'bestAccuracy':a,'bestKernel':k,'bestC':c,'bestGamma':g}]

pd.DataFrame(result)
```
